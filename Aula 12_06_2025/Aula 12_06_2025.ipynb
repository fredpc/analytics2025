```python
import pandas as pd
import numpy as np

# Supondo que seu DataFrame já foi carregado como df

# Criar a coluna 'Deck' a partir da primeira letra da cabine
df['Deck'] = df['Cabin'].apply(lambda x: x[0] if pd.notna(x) else np.nan)

# Agrupar por Deck, Embarked e Sex
agrupado = df.groupby(['Deck', 'Embarked', 'Sex']).agg(
    total=('Survived', 'count'),
    nao_sobreviventes=('Survived', lambda x: (x == 0).sum())
).reset_index()

# Filtrar apenas grupos com pelo menos 1 passageiro
agrupado = agrupado[agrupado['total'] > 0]

# Calcular a porcentagem de não sobreviventes
agrupado['porcentagem_nao_sobreviventes'] = 100 * agrupado['nao_sobreviventes'] / agrupado['total']

# Encontrar a terna com maior porcentagem de não sobreviventes
idx_max = agrupado['porcentagem_nao_sobreviventes'].idxmax()
terna_max = agrupado.loc[idx_max, ['Deck', 'Embarked', 'Sex']]
porcentagem_max = agrupado.loc[idx_max, 'porcentagem_nao_sobreviventes']

print(f"A terna com maior % de não sobreviventes é Deck={terna_max['Deck']}, Embarked={terna_max['Embarked']}, Sex={terna_max['Sex']}")
print(f"Com {porcentagem_max:.2f}% de não sobreviventes\n")

# Mostrar o total, mortos e porcentagem para essa terna, com proteção contra divisão por zero
df_filtrado = df[
    (df['Deck'] == terna_max['Deck']) &
    (df['Embarked'] == terna_max['Embarked']) &
    (df['Sex'] == terna_max['Sex'])
]

total = len(df_filtrado)
mortos = (df_filtrado['Survived'] == 0).sum()

if total > 0:
    porcentagem = mortos / total * 100
    print(f"Detalhes da terna selecionada:")
    print(f"Total de passageiros: {total}")
    print(f"Mortos: {mortos}")
    print(f"Porcentagem de não sobreviventes: {porcentagem:.2f}%")
    print("\nExemplos de passageiros dessa terna:")
    print(df_filtrado[['Survived', 'Sex', 'Cabin', 'Embarked']].head())
else:
    print("Nenhum passageiro encontrado nessa terna.")
```

```python
import pandas as pd

# Carrega o dataset (ajuste o caminho conforme necessário)
df = pd.read_csv('titanic_train.csv')

# Extrair a letra do Deck a partir da coluna Cabin (primeira letra)
df['Deck'] = df['Cabin'].astype(str).str[0]

# Filtrar dados com informações válidas para Deck, Embarked e Sex
df_filtrado = df[['Deck', 'Embarked', 'Sex', 'Survived']].dropna()

# Agrupar por Deck, Embarked e Sexo
agrupado = df_filtrado.groupby(['Deck', 'Embarked', 'Sex'])

# Contar total e número de sobreviventes
resultado = agrupado['Survived'].agg(
    total='count',
    survived='sum'
)

# Calcular porcentagem de não sobreviventes
resultado['not_survived_pct'] = ((resultado['total'] - resultado['survived']) / resultado['total']) * 100

# Ordenar do maior para o menor
resultado_ordenado = resultado.sort_values(by='not_survived_pct', ascending=False)

# Exibir o topo
print(resultado_ordenado.head(1))
```

```python
# Extrai a letra do deck (primeira letra da cabine)
df["Deck"] = df["Cabin"].astype(str).str[0]

# Filtra apenas colunas relevantes
df_filtrado = df[['Deck', 'Embarked', 'Sex', 'Survived']].dropna()

# Agrupa pelas ternas e calcula total e mortos
agrupado = df_filtrado.groupby(['Deck', 'Embarked', 'Sex'])

# Calcula total e mortos por grupo
estatisticas = agrupado['Survived'].agg(
    total='count',
    mortos=lambda x: (x == 0).sum()
)

# Adiciona coluna com porcentagem de não sobreviventes
estatisticas['%_mortos'] = estatisticas['mortos'] / estatisticas['total'] * 100

# Ordena do maior para o menor
resultado = estatisticas.sort_values('%_mortos', ascending=False)

# Mostra a terna com maior % de mortos
print("Terna com maior % de não sobreviventes:")
print(resultado.head(1))
```
