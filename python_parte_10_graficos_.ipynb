```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import scipy

from statsmodels.graphics.mosaicplot import mosaic

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.image as mpimg

import seaborn as sns

from sklearn import datasets

%matplotlib inline
```

<h3>Os datasets</h3>

Para essa primeira parte ilustraremos códigos de gráficos de barra, linha, dispersão, além de histogramas e boxplots utilizando os pacotes mais famosos Matplotlib, Seaborn, Plotly e Bokeh. Os datasets escolhidos para esse trabalho serão o Titanic (df) e Iris (iris).

<h4>Titanic</h4>

```python
df = pd.read_csv('assets/titanic_train.csv')
df = df[['Sex','Age','Fare','Embarked','Survived']]
df.Age.fillna(df.Age.mean(),inplace=True)
df.Embarked.fillna('S',inplace=True)
df['Survived'] = df['Survived'].map({0:'Died',1:'Survived'})
```

<h4>Iris</h4>

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
iris = pd.DataFrame(X, columns=iris.feature_names)
iris['Species'] = y
```

<h2>Gráfico de linha</h2>

<h3>Gráfico de linha no matplotlib</h3>

```python
result = df.sort_values(by=['Age'])

plt.figure(figsize=(8,4))

plt.plot(result['Age'],result['Fare'])

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Gráfico de linha")

plt.show()
```

<h2>Gráfico de barras</h2>

<h3>Exemplos de gráficos de barra no matplotlib</h3>

Bom para visualizações de dados categóricos. No nosso dataset temos dois tipos de dados categóricos: Sex (sexo) e Embarked (Portão de embarque).

```python
df['Embarked'].value_counts()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
embarked = ('S', 'C', 'Q')
y_pos = np.arange(len(embarked))
performance = [646,168,77]

ax.barh(y_pos, performance, align='center',color='green')

ax.set_yticks(y_pos)
ax.set_yticklabels(embarked)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Número de pessoas')
ax.set_ylabel('Portões')
ax.set_title('Por qual portão passaram os tripulantes?')

plt.show()
```

Uma alternativa ao gráfico de barras é o gráfico de pizza.

```python
labels = ('S', 'C', 'Q')
sizes = [646,168,77]
explode = (0.1, 0, 0)  # only "explode" the 1nd slice (i.e. 'S')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

No entanto, não se recomenda gráficos de pizza devido a dificuldade do ser humano de enxergar ângulos. Uma opção ao pizza, é o Donut.

```python
# create data
size = df["Survived"].value_counts().values
labels = df["Survived"].value_counts().index
 
# Create a circle for the center of the plot
my_circle=plt.Circle( (0,0), 0.6, color='white')

# Give color names
plt.pie(size, labels=labels, colors=['#EF476F','#06D6A0'])
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

Um exemplo de donut em análise bivariada.

```python
group_names=['Sobreviventes', 'Não sobreviventes']
group_size=[342,549]
subgroup_names=['Homens', 'Mulheres', 'Homens', 'Mulheres']
subgroup_size=[109,233,468,81]
 
# Create colors
a, b=[plt.cm.Greens, plt.cm.Reds] # https://matplotlib.org/stable/tutorials/colors/colormaps.html
 
# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(group_size, radius=1.3, labels=group_names, colors=[a(0.6), b(0.6)] )
plt.setp( mypie, width=0.3, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(subgroup_size, radius=1.3-0.3, labels=subgroup_names, labeldistance=0.7, colors=[a(0.4), a(0.2), b(0.5), b(0.4)])
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)
 
# show it
plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

x = np.arange(3)
males = [441,95,41]
females = [205,73,36]
width = 0.25

ax.bar(x, males, width, color='#3a0088', label="Masculino")
ax.bar(x + width, females, width, color='#930077', label="Feminino")

ax.set_xticks(x + width)
ax.set_xticklabels(['S', 'C', 'Q'])
ax.set_xlabel("Portões")
ax.set_ylabel("Número de pessoas")
ax.set_title("Por qual portão passaram os tripulantes?")
ax.legend(loc = 'upper right')

plt.show()
```

```python
plt.rcdefaults()

N = 3
males = [441,95,41]
females = [205,73,36]
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, males, width, color='#3a0088')
p2 = plt.bar(ind, females, width, color='#930077', bottom=males)

plt.ylabel('Número de pessoas')
plt.title('Por qual portão passaram os tripulantes?')
plt.xticks(ind, ('S', 'C', 'Q'))
plt.yticks(np.arange(0, 700, 100))
plt.legend((p1[0], p2[0]), ('Masculino', 'Feminino'))

plt.show()
```

<h2>Gráficos de dispersão</h2>

<h3>Gráfico de dispersão com o matplotblib</h3>

```python
plt.rcdefaults()

colors = 50*['red']+50*['green']+50*['blue']
area = iris['petal length (cm)']*15

xdata = iris['sepal length (cm)']
ydata = iris['sepal width (cm)']

plt.scatter(x=xdata, y=ydata, s=area, c=colors, alpha=0.5)
plt.title('Iris data')
plt.xlabel('Sepal length (cm)')
plt.ylabel('Sepal width (cm)')

# anotações: https://matplotlib.org/users/annotations.html#annotating-with-text-with-box
bbox_props_setosa = dict(boxstyle="square,pad=0.1", fc="red", ec="r", lw=2, alpha=0.4)
plt.annotate('setosa',xy=(4.5,3.7),size=10,bbox=bbox_props_setosa)
bbox_props_virginica = dict(boxstyle="square,pad=0.1", fc="blue", ec="b", lw=2, alpha=0.4)
plt.annotate('virginica',xy=(7.25,3.4),size=10, bbox=bbox_props_virginica)
bbox_props_versicolor = dict(boxstyle="square,pad=0.1", fc="green", ec="g", lw=2, alpha=0.4)
plt.annotate('versicolor',xy=(5.3,2.1),size=10, bbox=bbox_props_versicolor)

plt.show()
```

<h3>Gráfico de dispersão com o Seaborn</h3>

```python
plt.rcdefaults()

iris['Species'] = 50*['Setosa'] + 50*['Virginica'] + 50*['Versicolor']

# Escolhendo as cores
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]

# criando gráfico para visualização de relações de 2 a 2 features
sns.pairplot(iris,
             hue='Species',
             palette=flatui,
             vars = [
                 'sepal length (cm)',
                 'sepal width (cm)',
                 'petal length (cm)',
                 'petal width (cm)'
             ],
             size=3,
             diag_kind="kde") # can be hist

plt.show()
```

<h3>Heatmap com Seaborn</h3>

Fonte: https://seaborn.pydata.org/generated/seaborn.heatmap.html

```python
plt.rcdefaults()

myCrosstable = pd.crosstab(df['Embarked'], df['Sex'])

ax = sns.heatmap(myCrosstable, linewidths=.5)
plt.show()
```

```python
# Fazer a média das características por espécie (opcional para simplificação visual)
mean_values = iris.groupby('Species').mean()

# Criar o heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(mean_values, annot=True, cmap="YlGnBu", cbar=True, fmt=".2f")
plt.title("Heatmap das características médias por espécie - Iris Dataset")
plt.xlabel("Características")
plt.ylabel("Espécies")
plt.tight_layout()
plt.show()
```

```python
plt.rcdefaults()
mean_crosstab_sex_embarked = pd.pivot_table(df, values='Fare', index=['Embarked'],columns=['Sex'], aggfunc=np.mean)
ax = sns.heatmap(mean_crosstab_sex_embarked, linewidths=.5 )
# cmap can be "BuPu" "Blues" "YlGnBu" "Greens"
ax.set_title("Idade média por portão de embarque e sexo")
plt.show()
```

```python
import random
E = []
for i in range(30):
    x = random.randint(1,51)
    y = random.randint(1,51)
    E.append((x,y))
```

```python
G=nx.DiGraph()
G.add_nodes_from(range(1,51))
G.add_edges_from(E)

plt.figure(figsize=(8,8))            
nx.draw_circular(G,node_color='g', edge_color='#909090', node_size=900)
plt.axis('equal')
plt.show()
```

    - O dashboard deve mostrar informações atreladas a tomadas de decisões

```python
import sqlite3
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import numpy as np
from datetime import datetime

# Paleta de cores profissional unificada
COLORS = {
    'primary': '#1f2937',      # Azul escuro
    'secondary': '#3b82f6',    # Azul
    'accent': '#10b981',       # Verde
    'warning': '#f59e0b',      # Laranja
    'danger': '#ef4444',       # Vermelho
    'purple': '#8b5cf6',       # Roxo
    'teal': '#14b8a6',         # Azul claro
    'pink': '#ec4899',         # Rosa
    'background': '#f8fafc',   # Cinza claro
    'text': '#1f2937'          # Texto escuro
}

# Paleta unificada para todos os gráficos (10 cores harmoniosas)
CHART_COLORS = ['#3b82f6', '#10b981', '#f59e0b', '#ef4444', '#8b5cf6', '#14b8a6', '#ec4899', '#6366f1', '#f97316', '#06b6d4']

# Conexão com o banco de dados
conn = sqlite3.connect('assets/chinook.db')

# Consulta 1: Vendas por Gênero
df_genres = pd.read_sql("""
    SELECT g.Name AS genre, SUM(ii.UnitPrice * ii.Quantity) AS sales
    FROM genres g
    JOIN tracks t ON g.GenreId = t.GenreId
    JOIN invoice_items ii ON t.TrackId = ii.TrackId
    GROUP BY g.Name
    ORDER BY sales DESC
    LIMIT 8
""", conn)

# Consulta 2: Vendas por País (CORRIGIDA - ordem decrescente)
df_countries = pd.read_sql("""
    SELECT c.Country AS country, SUM(i.Total) AS sales
    FROM customers c
    JOIN invoices i ON c.CustomerId = i.CustomerId
    GROUP BY c.Country
    ORDER BY sales DESC
    LIMIT 10
""", conn)

# Consulta 3: Vendas por Tipo de Mídia
df_media = pd.read_sql("""
    SELECT m.Name AS media_type, SUM(ii.UnitPrice * ii.Quantity) AS sales
    FROM media_types m
    JOIN tracks t ON m.MediaTypeId = t.MediaTypeId
    JOIN invoice_items ii ON t.TrackId = ii.TrackId
    GROUP BY m.Name
    ORDER BY sales DESC
""", conn)

# Consulta 4: Top 8 Clientes
df_customers = pd.read_sql("""
    SELECT c.FirstName || ' ' || c.LastName AS customer, 
           SUM(i.Total) AS total_spent,
           COUNT(i.InvoiceId) AS orders_count
    FROM customers c
    JOIN invoices i ON c.CustomerId = i.CustomerId
    GROUP BY c.CustomerId
    ORDER BY total_spent DESC
    LIMIT 8
""", conn)

# Consulta 5: Vendas mensais (Timeline) - CORRIGIDA
df_timeline = pd.read_sql("""
    SELECT 
        strftime('%Y-%m', i.InvoiceDate) AS month,
        SUM(i.Total) AS monthly_sales,
        COUNT(i.InvoiceId) AS orders_count
    FROM invoices i
    GROUP BY strftime('%Y-%m', i.InvoiceDate)
    ORDER BY month ASC
""", conn)

conn.close()

# Ordenando países do maior para o menor (para exibir corretamente no gráfico horizontal)
df_countries = df_countries.sort_values('sales', ascending=True)  # Invertido para barras horizontais

# Criando o dashboard com subplots
fig = make_subplots(
    rows=3, cols=2,
    subplot_titles=[
        '🎵 Distribuição de Vendas por Gênero Musical',
        '🌍 Performance de Vendas por País',
        '💿 Análise por Tipo de Mídia (Treemap)',
        '👥 Top Clientes por Faturamento',
        '📈 Evolução das Vendas Mensais',
        '🎯 Métricas de Engajamento'
    ],
    specs=[
        [{"type": "pie"}, {"type": "bar"}],
        [{"type": "treemap"}, {"type": "bar"}],
        [{"type": "scatter", "colspan": 2}, None]
    ],
    vertical_spacing=0.12,
    horizontal_spacing=0.1
)

# 1. Gráfico de Pizza - Vendas por Gênero
fig.add_trace(
    go.Pie(
        labels=df_genres['genre'],
        values=df_genres['sales'],
        hole=0.4,
        textinfo='label+percent',
        textposition='outside',
        marker=dict(
            colors=CHART_COLORS[:len(df_genres)],
            line=dict(color='white', width=3)
        ),
        hovertemplate='<b>%{label}</b><br>' +
                     'Vendas: $%{value:,.2f}<br>' +
                     'Participação: %{percent}<br>' +
                     '<extra></extra>'
    ),
    row=1, col=1
)

# 2. Gráfico de Barras Horizontais - Vendas por País
fig.add_trace(
    go.Bar(
        y=df_countries['country'],
        x=df_countries['sales'],
        orientation='h',
        marker=dict(
            color=CHART_COLORS[:len(df_countries)],
            line=dict(color='white', width=1)
        ),
        text=[f'${x:,.0f}' for x in df_countries['sales']],
        textposition='inside',
        hovertemplate='<b>%{y}</b><br>' +
                     'Vendas: $%{x:,.2f}<br>' +
                     '<extra></extra>'
    ),
    row=1, col=2
)

# 3. Treemap - Vendas por Mídia
fig.add_trace(
    go.Treemap(
        labels=df_media['media_type'],
        values=df_media['sales'],
        parents=[""] * len(df_media),
        textinfo="label+value+percent parent",
        marker=dict(
            colors=CHART_COLORS[2:2+len(df_media)],
            line=dict(color='white', width=2)
        ),
        hovertemplate='<b>%{label}</b><br>' +
                     'Vendas: $%{value:,.2f}<br>' +
                     'Participação: %{percentParent}<br>' +
                     '<extra></extra>'
    ),
    row=2, col=1
)

# 4. Gráfico de Barras Verticais - Top Clientes
fig.add_trace(
    go.Bar(
        x=df_customers['customer'],
        y=df_customers['total_spent'],
        marker=dict(
            color=CHART_COLORS[4:4+len(df_customers)],
            line=dict(color='white', width=2)
        ),
        text=[f'${y:,.0f}' for y in df_customers['total_spent']],
        textposition='outside',
        customdata=df_customers['orders_count'],
        hovertemplate='<b>%{x}</b><br>' +
                     'Total Gasto: $%{y:,.2f}<br>' +
                     'Pedidos: %{customdata}<br>' +
                     '<extra></extra>'
    ),
    row=2, col=2
)

# 5. Gráfico de Linha com Área - Timeline de Vendas
fig.add_trace(
    go.Scatter(
        x=df_timeline['month'],
        y=df_timeline['monthly_sales'],
        mode='lines+markers',
        line=dict(color=COLORS['secondary'], width=4),
        marker=dict(
            size=8,
            color=COLORS['accent'],
            line=dict(color='white', width=2)
        ),
        fill='tonexty',
        fillcolor='rgba(59, 130, 246, 0.1)',
        hovertemplate='<b>%{x}</b><br>' +
                     'Vendas: $%{y:,.2f}<br>' +
                     '<extra></extra>'
    ),
    row=3, col=1
)

# Atualizando layout dos eixos
fig.update_xaxes(showgrid=True, gridcolor='rgba(0,0,0,0.1)', row=1, col=2)
fig.update_yaxes(showgrid=True, gridcolor='rgba(0,0,0,0.1)', row=1, col=2)

# Eixos específicos para gráfico de clientes
fig.update_xaxes(showgrid=True, gridcolor='rgba(0,0,0,0.1)', row=2, col=2)
fig.update_yaxes(showgrid=True, gridcolor='rgba(0,0,0,0.1)', row=2, col=2)

fig.update_xaxes(title_text="Período", showgrid=True, gridcolor='rgba(0,0,0,0.1)', row=3, col=1)
fig.update_yaxes(title_text="Vendas Mensais ($)", showgrid=True, gridcolor='rgba(0,0,0,0.1)', row=3, col=1)

# Configurações gerais do layout
fig.update_layout(
    title=dict(
        text='<b>🎵 CHINOOK MUSIC STORE - DASHBOARD EXECUTIVO</b>',
        x=0.5,
        font=dict(size=28, color=COLORS['text']),
        pad=dict(b=30)
    ),
    height=1200,
    showlegend=False,
    plot_bgcolor='white',
    paper_bgcolor=COLORS['background'],
    font=dict(family="Arial, sans-serif", size=12, color=COLORS['text']),
    margin=dict(l=50, r=50, t=100, b=50)
)

# Atualizando títulos dos subplots
for i, annotation in enumerate(fig['layout']['annotations']):
    annotation['font'] = dict(size=16, color=COLORS['text'])
    annotation['bgcolor'] = 'white'
    annotation['bordercolor'] = COLORS['primary']
    annotation['borderwidth'] = 2
    annotation['borderpad'] = 10

# Exibindo o dashboard
fig.show()

# Calculando métricas estratégicas
total_sales = df_genres['sales'].sum()
top_genre = df_genres.iloc[0]
top_country = df_countries.iloc[-1]  # Último da lista ordenada (maior valor)
top_media = df_media.iloc[0]
top_customer = df_customers.iloc[0]

# Restante do código para insights estratégicos
print("=" * 80)
print("🎯 RELATÓRIO ESTRATÉGICO - RECOMENDAÇÕES PARA AÇÃO")
print("=" * 80)
print("\n📌 DECISÕES PRIORITÁRIAS:")
print(f"1. ALOCAÇÃO DE ESTOQUE: Aumentar em 30% o catálogo de {top_genre['genre']} (gera ${top_genre['sales']:,.0f}/ano)")
print(f"   → AÇÃO: Contratar mais artistas do gênero {top_genre['genre']} e criar playlists temáticas")

print(f"\n2. EXPANSÃO GEOGRÁFICA: Priorizar campanhas em {top_country['country']} (${top_country['sales']:,.0f} em vendas)")
print(f"   → AÇÃO: Traduzir site para o idioma local e oferecer frete grátis nesta região")

print(f"\n3. FORMATOS DE MÍDIA: Investir em {top_media['media_type']} (${top_media['sales']:,.0f} faturados)")
print(f"   → AÇÃO: Reduzir estoque de mídias físicas e ampliar parcerias com streamers")

print(f"\n4. PROGRAMA DE FIDELIDADE: Criar benefícios exclusivos para {top_customer['customer']} (gastou ${top_customer['total_spent']:,.0f})")
print(f"   → AÇÃO: Oferecer desconto vitalício de 15% e early access a lançamentos")

print("\n💡 PRÓXIMOS PASSOS RECOMENDADOS:")
print("- Testar bundle de gêneros menos populares com os top 3")
print("- Criar pacote 'Descoberta Musical' para novos clientes")
print("- Desenvolver relatório semanal de mix de vendas por país")
print("=" * 80)
```
